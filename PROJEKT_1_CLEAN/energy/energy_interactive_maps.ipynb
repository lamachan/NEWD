```python
import pandas as pd
import plotly.graph_objects as go
```

```python
df = pd.read_csv('global-data-on-sustainable-energy.csv')
df.head()
```

```python
def plot_world_map_with_slider(column_name, file_name):
    fig = go.Figure()
    for year in range(2000, 2021):
        filtered_df = df[df['Year'] == year]
        trace = go.Choropleth(
            locations=filtered_df['Entity'],
            z=filtered_df[column_name],
            locationmode='country names',
            colorscale='Electric',
            zmin=df[column_name].min(),
            zmax=df[column_name].max(),
            visible=False
        )
        fig.add_trace(trace)

    fig.data[0].visible = True
    steps = []
    for i in range(len(fig.data)):
        step = dict(
            method='update',
            args=[{'visible': [False] * len(fig.data)},
                  {'title_text': f'{column_name} Map - {2000 + i}', 'frame': {'duration': 1000, 'redraw': True}}],
            label=str(2000 + i)
        )
        step['args'][0]['visible'][i] = True
        steps.append(step)

    sliders = [dict(
        active=0,
        steps=steps,
        currentvalue={"prefix": "Year: ", "font": {"size": 14}},
    )]

    fig.update_layout(
        title_text=f'{column_name} Map with slider',
        title_font_size=24,
        title_x=0.5,
        geo=dict(
            showframe=True,
            showcoastlines=True,
            projection_type='natural earth'
        ),
        sliders=sliders,
        height=500,
        width=1000,
        font=dict(family='Arial', size=12),
        margin=dict(t=80, l=50, r=50, b=50),
    )
    fig.show()
    # Export as HTML
    fig.write_html(file_name)
```

```python
column_name = 'Access to electricity (% of population)'
file_name = 'access_to_electricity_map.html'
plot_world_map_with_slider(column_name, file_name)
```

```python
column_name = 'Renewable energy share in the total final energy consumption (%)'
file_name = 'renewable_energy_map.html'
plot_world_map_with_slider(column_name, file_name)
```
